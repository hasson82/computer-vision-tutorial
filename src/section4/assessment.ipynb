```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# open image
img = cv2.imread('..\\..\\data\\jj_and_omer.jpeg')
type(img)
```

```python
plt.imshow(img)
```

```python
fixed_img = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
plt.imshow(fixed_img)
```

```python
plt.imshow(fixed_img)
```

```python
cv2.rectangle(fixed_img, pt1=(270, 410), pt2=(405, 540), color=(255, 0, 0), thickness=10)
cv2.circle(fixed_img, center=(200, 150), radius=145, color=(0, 255, 0), thickness=10)
```

```python
#flip image
flip_img = cv2.flip(fixed_img, 1)
flipX2_img = cv2.flip(flip_img, 0)
plt.imshow(flipX2_img)
```
